```python
concrete_data.rename(columns={"Cement (component 1)(kg in a m^3 mixture)":'Cement',
                               "Blast Furnace Slag (component 2)(kg in a m^3 mixture)":'Blast_Furnace_Slag',
                               "Fly Ash (component 3)(kg in a m^3 mixture)":'Fly_Ash', "Water  (component 4)(kg in a m^3 mixture)": 'Water',
                               "Superplasticizer (component 5)(kg in a m^3 mixture)":'Superplasticizer',
                                "Coarse Aggregate  (component 6)(kg in a m^3 mixture)":'Coarse-Aggregate',
                               "Fine Aggregate (component 7)(kg in a m^3 mixture)": 'Fine_Aggregate',
                                "Age (day)": "Age_day",
                        "Concrete_target ": 'Concrete_target'
                                
                               }, inplace=True )
```

encoding features### 

```python
blast_mean = np.mean(concrete_data.Blast_Furnace_Slag)
fly_mean = np.mean(concrete_data.Fly_Ash)
superplastic_mean = np.mean(concrete_data.Superplasticizer)
```

```python
concrete_data.Blast_Furnace_Slag=concrete_data.Blast_Furnace_Slag.mask(concrete_data.Blast_Furnace_Slag==0).fillna(blast_mean)
```

```python
concrete_data.Fly_Ash=concrete_data.Fly_Ash.mask(concrete_data.Fly_Ash==0).fillna(fly_mean)
```

```python
concrete_data.Superplasticizer=concrete_data.Superplasticizer.mask(concrete_data.Superplasticizer==0).fillna(superplastic_mean)
```

```python
y = concrete_data['Concrete compressive strength(MPa, megapascals) ']
```

```python
x = concrete_data.drop(columns='Concrete compressive strength(MPa, megapascals) ')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2)
```

```python
X_train.shape
```

```python
X_train.size
```

```python
X_val.shape
```

```python
X_test.shape
```

```python
# Data Normalization
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

X_train -= mean
X_train /= std

X_test -= mean
X_test /= std


X_val -= mean
X_val /= std
```

```python
X_train.head()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
network = Sequential()
```

```python
network.add(layers.Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
network.add(layers.Dense(8, activation='relu'))
network.add(layers.Dense(6))
```

```python
network.compile(optimizer='adam', loss='mse', metrics='mae')
```

```python
history = network.fit(X_train, Y_train, batch_size=30, verbose=0, epochs=100, validation_data=(X_val, Y_val))
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, 101)
```

```python
print('Training Loss: ' + str(train_loss[-1]))
print('*'*20)
print('Validation Loss' + str(val_loss[-1]))
```

```python
plt.plot(epochs, train_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Val loss')
plt.title('Train and Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

refitting the model### 

```python
network.add(layers.Dense(6, activation='relu', input_shape=(X_train.shape[1],)))
network.add(layers.Dense(4, activation='relu'))
network.add(layers.Dense(1))
```

```python
network.compile(optimizer='adam', loss='mse', metrics='mae')
```

```python
history = network.fit(X_train, Y_train, batch_size=30, verbose=0, epochs=200, validation_data=(X_val, Y_val))
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, 201)
```

```python
print('Training Loss: ' + str(train_loss[-1]))
print('*'*20)
print('Validation Loss' + str(val_loss[-1]))
```

```python
plt.plot(epochs, train_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Val loss')
plt.title('Train and Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
prediction = network.predict(X_test)
```

```python
network.evaluate(X_test, Y_test)
```
